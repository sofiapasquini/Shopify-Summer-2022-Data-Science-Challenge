```python
#import the relevant packages
import pandas as pd
import numpy as np
```

```python
#import the data set
data=pd.read_csv("Dat_Sci_Challenge.csv")
data.head(100)
```

```python
#How many times does the outlier transaction occur? Use the order_amount value
outlier_idx=np.where(data.order_amount>=704000)
outlier_count=np.size(outlier_idx)
print("the number of outlier transactions: ", outlier_count)
```

```python
#How many transactions are in the entire dataset? How many of these are outliers?
print("total transactions: ",data.shape[0])
print("outlier proportion of total transactions: ", np.round((outlier_count/data.shape[0])*100, decimals=2))
```

```python
#Remove the outliers from the dataset
outlier_idx=np.asarray(outlier_idx)
outlier_idx=outlier_idx.reshape(17,).tolist()
data_clean=data.drop(outlier_idx, axis=0)
```

```python
#Now calculate the median order_amount value
print("median on 'cleaned' dataset: ", np.median(data_clean.order_amount))
```
